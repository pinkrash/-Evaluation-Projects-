```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.3,random_state=21)
```

```python
models=[ LinearRegression(),KNeighborsRegressor(), DecisionTreeRegressor(),Ridge(),RandomForestRegressor()]
for i in models:
    i.fit(xtrain,ytrain)
    pred=i.predict(xtest)
    
    print(i)
    print(mean_squared_error(ytest,pred))
    print(mean_absolute_error(ytest,pred))
    print(np.sqrt(mean_squared_error(ytest,pred)))
    print(r2_score(ytest,pred))
    print('=='*50)
```

# REGION PREDICTION

```python
df=pd.read_csv('avacado.csv')
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
```

```python
df['region'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
lb= LabelEncoder()
df['type']=lb.fit_transform(df['type'])
```

```python
df
```

```python
skew_list=['Total Volume','4046','4225','4770','Total Bags','Small Bags','Large Bags','XLarge Bags']
for i in skew_list:
    df[i]=boxcox(df[i],0.5)
```

```python
df.drop(['Date'],axis=1,inplace=True)
```

```python
df.drop(['Unnamed: 0','Total Bags'],axis=1,inplace=True)
```

```python
df
```

```python
def first_letter(letter):
    regions = df[df['region'].str.startswith(letter)]
    for word in regions['region']:
        return word
```

```python
def letter(s):
    first_letter = s[0].upper()
    if first_letter == 'A':
        return 1
    elif first_letter == 'B':
        return 2
    elif first_letter == 'C':
        return 3
    elif first_letter == 'D':
        return 4
    elif first_letter == 'G':
        return 5
    elif first_letter == 'H':
        return 6
    elif first_letter == 'I':
        return 7
    elif first_letter == 'J':
        return 8
    elif first_letter == 'L':
        return 9
    elif first_letter == 'M':
        return 10
    elif first_letter == 'N':
        return 11
    elif first_letter == 'O':
        return 12
    elif first_letter == 'P':
        return 13
    elif first_letter == 'R':
        return 14
    elif first_letter == 'S':
        return 15
    elif first_letter == 'T':
        return 16
    elif first_letter == 'W':
        return 17
    else:
        return 0
    
```

```python
df['region'] = df['region'].apply(letter)
df
```

```python
z=np.abs(zscore(df).values)
threshold=3
print(np.where(z>3))
```

```python
from sklearn.svm import SVC

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X=df[['AveragePrice','Total Volume','4046','4225','4770','Small Bags','Large Bags','XLarge Bags','type','year']]
X
```

```python
Y=df['region']
Y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=.3,random_state=21)
```

```python
models=[SVC(), DecisionTreeClassifier(),RandomForestClassifier()]
for i in models:
    i.fit(xtrain,ytrain)
    pred=i.predict(xtest)
    
    print(i)
    print(accuracy_score(ytest,pred))
    print('=='*50)
```
